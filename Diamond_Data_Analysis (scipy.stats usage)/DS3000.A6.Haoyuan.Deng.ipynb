## Question 7: 
(10 pts) Using a 2-sample t-test**, determine if there is a statistical difference between the price of diamonds with a cut that is considered:
'Good' vs 'Very Good'
'Premium' vs 'Ideal'
**set alpha to 5% i.e. 0.05.

```python
# Filtering data for the comparisons
good_price = diamond_df[diamond_df['cut'] == 'Good']['price']
very_good_price = diamond_df[diamond_df['cut'] == 'Very Good']['price']

premium_price = diamond_df[diamond_df['cut'] == 'Premium']['price']
ideal_price = diamond_df[diamond_df['cut'] == 'Ideal']['price']

alpha = 0.05

# T-tests
t_stat, p_value = stats.ttest_ind(good_price, very_good_price)
t_stat2, p_value2 = stats.ttest_ind(premium_price, ideal_price)

print(f"T-Test (Good vs Very Good): p-value = {p_value:.3f}")
print(f"T-Test (Premium vs Ideal): p-value = {p_value2:.3f}")

# Compare the p-values with alpha to determine significance
if p_value < alpha:
    print('There is a statistically difference between Good and Very Good cuts.')
else:
    print('There is no statistically difference between Good and Very Good cuts.')

if p_value2 < alpha:
    print('There is a statistically difference between Premium and Ideal cuts.')
else:
    print('There is no statistically difference between Premium and Ideal cuts.')
```

Good vs Very Good: Since the p-value (0.4056) > 0.05, we fail to reject the null hypothesis, meaning there is no statistically significant difference in diamond prices between the two cuts. This suggests that diamonds with a Good cut and a Very Good cut have similar price distributions.

Premium vs Ideal: Since the p-value (0.0000) < 0.05, we reject the null hypothesis, meaning there is a statistically significant difference in diamond prices between these two cuts. This indicates that Premium-cut diamonds are priced differently from Ideal-cut diamonds.

```python
# Apply log transformation
diamond_df['log_price'] = np.log(diamond_df['price'])
```

```python
# Create a boxplot for log-transformed price distribution by cut
plt.figure(figsize=(12,6))
sns.boxplot(x='cut', y='log_price', data=diamond_df)
plt.xlabel("Cut")
plt.ylabel("Log Price")
plt.title("Log-Transformed Price Distribution by Cut")
plt.show()
```

In the original data, there are many high-price outliers extending far above the upper whiskers of the boxplots. In the transformed data, the extreme outliers are reduced, and prices appear more evenly spread

3. Repeat the statistical tests in question 7 using the corrected data. Are there any changes in the statistical significance? Explain the results.

```python
# Extract log-transformed prices the two cuts comparisons
good_log_prices = diamond_df[diamond_df['cut'] == 'Good']['log_price']
very_good_log_prices = diamond_df[diamond_df['cut'] == 'Very Good']['log_price']

premium_log_prices = diamond_df[diamond_df['cut'] == 'Premium']['log_price']
ideal_log_prices = diamond_df[diamond_df['cut'] == 'Ideal']['log_price']

# Perform a 2-sample t-test both comparisons with log-transformed data
t_stat_log, p_value_log = stats.ttest_ind(good_log_prices, very_good_log_prices)
t_stat_log2, p_value_log2 = stats.ttest_ind(premium_log_prices, ideal_log_prices)

print(f"T-statistic (log-transformed): {t_stat_log:.3f}, P-value (log-transformed): {p_value_log:.3f}")
print(f"T-statistic (log-transformed): {t_stat_log2:.3f}, P-value (log-transformed): {p_value_log2:.3f}")
```

Good vs. Very Good Cut:

Before transformation: p = 0.419 (not significant, p > 0.05)

After transformation: p = 0.0107 (significant, p < 0.05)

Changes: The log transformation corrected the skewness, revealing a statistically significant difference that was not previously detected. The raw price data had high variability, which may have hid the difference between these two groups.

Premium vs. Ideal Cut:

Before transformation: p  0 (significant)

After transformation: p  0 (still significant)

Changes: The significance remained strong even after correcting for skewness, confirming that the price difference between "Premium" and "Ideal" cuts is very real and substantial.
